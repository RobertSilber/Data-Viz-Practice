```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### PART 1 ###

```python
atomic_weights = [1, 4, 7, 9, 10, 12]
labels = ['hydrogen', 'helium', 'lithium', 'beryllium', 'boron', 'carbon']
explode1 = [0.2, 0, 0, 0, 0, 0]
explode2 = [0, 0.2, 0, 0, 0, 0]
weight_array = np.array(atomic_weights)
fig = plt.figure(figsize=(4,3), dpi=144)

ax1 = fig.add_subplot(121)
ax1.pie(atomic_weights, explode=explode1, labels=labels, autopct='%1.1f%%')



ax2 = fig.add_subplot(122)
ax2.pie(atomic_weights, explode=explode2, labels=labels,autopct= lambda x: '{:.0f}'.format(x*weight_array.sum()/100))
```

### PART 3 ###

```python
days_list = ['January 3rd 2023', 'January 5th 2023', 'January 7th 2023',
             'January 9th 2023', 'January 11th 2023', 'January 13th 2023',
            'January 15th 2023', 'January 17th 2023']

floats = np.random.uniform(low=100, high=200, size=8)

P3_df = pd.DataFrame(list(zip(days_list,floats)))
P3_df.columns = ['date', 'value']
P3_df['date'] = pd.to_datetime(P3_df['date'])
print(P3_df.dtypes)
P3_df.set_index('date')
```

```python
fig, axes = plt.subplots(1,2, figsize=(25,4))
axes[0].plot(P3_df['date'], P3_df['value'])
axes[0].set_title('Line Plot of Values v Days')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Value')

axes[1].bar(P3_df['date'], P3_df['value'])
axes[1].set_title('Bar Plot of Values v Days')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Value')
```
